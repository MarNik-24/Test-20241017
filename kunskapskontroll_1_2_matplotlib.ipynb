### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1. An explicit "Axes" or "object-oriented" interface approach: by using (fig, ax = plt.subplots()).
In this method we creat figures and axes explicity (each object is explicitly referenced, and used to make the next object). 

2. An implicit "pyplot" interface approach: by using (plt.plot()).
In this method, Matplotlib manages the figure and axes implicitly.

For simple plots, we can use implicit pyplot method, but the recommended way is explicit method especially with complicated or multiple plots. Since, explicit method give us more flexibility and allows us to customize the objects after they are created.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: The small window or canvas which holds one or multiple plot. 

Axes (subplots): Axes is a plot that is placed on a figure and usually has a pair of axis (x-axis & y-axis).

Axis: Axis is any of the lines on a Axes (icluding information on the scale, limits, ticks and ticklabels).

Artist: Almost everything we see on the figure, including lines & texts.

### When plotting in Matplotlib, what is the expected input data type?

numpy.array or numpy.ma.masked_array or objects that can be passed to numpy.asarray are the expected input data types. We can also use 'array-like' data such as pandas and numpy.matrix after converting them to numpy.array. 

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4) # linspace returns evenly spaced numbers over a specified interval(-4 to 4)
y = x**2

# The pyplot approach
plt.plot(x ,y)
plt.title('y = x^2')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
x = np.linspace(-4, 4) 
y = x**2

# The object-oriented approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('y = x^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# I use the object-oriented approach
# I Create a figure with 2 axes (ax1 will be the scatter plot & ax2 will be the bar plot) 
# Numbers in the parentheses determine the number of subplots (1 row with two columns). figsize determines the size of the figure.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =[10, 5])  

ax1.scatter(x, y) # Creates a scatter plot on the first axes using provided x & y values
ax1.set_title('y = x^2')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.bar(names, values) # Creates a bar plot on the second axes using names & values
ax2.set_title('Fruits')
ax2.set_xlabel('Name of friuts')
ax2.set_ylabel('Vlues')
```
